### 2001年轉成CSV檔

```python
def worldGame2001(path:str)->pd.DataFrame:
    file = open(path, "r")
    content = file.read()
    file.close()

    #僅保留需要的欄位
    s = 'old bwt Team sq bp dl Total points'

    lst_cols = []
    dic_cols = {}
    for col in s.split(' '):
        lst_cols.append(col)
        dic_cols[col] = []

    lst_cols.append('wc')
    dic_cols['wc'] = []

    #將不同量集分開
    lst = content.split('\n')
    dic_lifters = {}

    curr_quantity = ''
    for lifter in lst[1:]:
        if 'Wom' in lifter or 'Men' in lifter:
            curr_quantity = lifter.lower()
            dic_lifters[curr_quantity]=[]
        else:
            #Disq 代表至少一項沒成績，不用加入
            if 'Disq' not in lifter:
                dic_lifters[curr_quantity].append(lifter)

    # #過濾掉不需要的資訊並轉成DataFrame
    for key in dic_lifters.keys():
        # print(key)
        ary = None
        for lifer in dic_lifters[key]:

            is_country = 0 
            i=0
            for s in lifer.split(' ')[1:]:
                try:
                    # print(s)
                    if ',' in s:
                        dic_cols[lst_cols[i]].append(float(s.replace(',','.')))
                        i+=1
                    else:
                        dic_cols[lst_cols[i]].append(float(s))
                        i+=1
                        if is_country == 0:
                            is_country = 1
                except:
                    if is_country == 1:
                        dic_cols[lst_cols[i]].append(s)
                        i+=1
                        is_country=2
                    None
            dic_cols[lst_cols[-1]].append(key)

    df = pd.DataFrame(data=dic_cols)
    df['old'] = 2001 - (df['old']+1900)
    df['old'] = df['old'].astype(int)
    df['year'] = 2001
    df = df.rename(columns={ col:col.lower() for col in df.columns})
    df = df[['year','wc','old', 'team', 'bwt', 'sq', 'bp', 'dl', 'total', 'points']]
    return df
```

```python
df = worldGame2001('./raw_data/2001.txt')
df.to_csv('./preprocess/worldgame_2001.csv')
df
```

### 將2005的比賽資料轉成 csv檔

```python
def worldGame2005(path:str,is_men:bool)->pd.DataFrame:

    file = open(path, "r")
    content = file.read()
    file.close()

    #僅保留需要的欄位
    s = 'old Team Bwt IPF_GL_Cf. Sq BP DL Total IPF_GL'

    lst_cols = []
    dic_cols = {}
    for col in s.split(' '):
        lst_cols.append(col)
        dic_cols[col] = []

    lst_cols.append('wc')
    dic_cols['wc'] = []

    #將不同量集分開
    lst = content.split('\n')
    dic_lifters = {}

    curr_quantity = ''
    for lifter in lst[1:]:
        if 'kg' in lifter:
            
            curr_quantity = lifter.replace(' - ','').replace(' ','').replace('kg','').replace(',','.')
            dic_lifters[curr_quantity]=[]
        else:
            #Disq 代表至少一項沒成績，不用加入
            if 'Disq' not in lifter:
                dic_lifters[curr_quantity].append(lifter)

    #過濾掉不需要的資訊並轉成DataFrame
    for key in dic_lifters.keys():
        # print(key)
        ary = None
        for lifer in dic_lifters[key]:

            is_country = 0 
            i=0
            for s in lifer.split(' ')[1:]:
                try:
                    # print(s)
                    if ',' in s:
                        dic_cols[lst_cols[i]].append(float(s.replace(',','.')))
                        i+=1
                    elif '#' in s: #代表成功後，加重後試舉失敗
                        dic_cols[lst_cols[i]].append(0)
                        i+=1
                    else:
                        dic_cols[lst_cols[i]].append(float(s))
                        i+=1
                        if is_country == 0:
                            is_country = 1
                except:
                    if is_country == 1:
                        dic_cols[lst_cols[i]].append(s)
                        i+=1
                        is_country=2
                    None
            dic_cols[lst_cols[-1]].append(key)
            
    df = pd.DataFrame(data=dic_cols)
    df['old'] = 2005 - df['old']
    df['old'] = df['old'].astype(int)
    df['year'] = 2005
    if is_men:
        df['gender'] = 'men'
    else:
        df['gender'] = 'women'
    df = df.rename(columns={ col:col.lower() for col in df.columns})
    df = df[['year','wc','gender','old', 'team', 'bwt', 'ipf_gl_cf.', 'sq', 'bp','dl', 'total', 'ipf_gl']]
    return df
```

```python
dfWomen = worldGame2005('./raw_data/2005women.txt',False)
dfMen = worldGame2005('./raw_data/2005men.txt',True)
df = pd.concat([dfWomen,dfMen])
df.to_csv('./preprocess/worldgame_2005.csv')
df
```

```python
def worldGame2009(path:str,is_men:bool)->pd.DataFrame:

    file = open(path, "r")
    content = file.read()
    file.close()

    #僅保留需要的欄位
    s = 'old Team Bwt IPF_GL_Cf. Sq BP DL Total IPF_GL Pts'

    lst_cols = []
    dic_cols = {}
    for col in s.split(' '):
        lst_cols.append(col)
        dic_cols[col] = []

    lst_cols.append('wc')
    dic_cols['wc'] = []

    #將不同量集分開
    lst = content.split('\n')
    dic_lifters = {}

    curr_quantity = ''
    for lifter in lst[1:]:
        if ' - MW' in lifter or ' - LW' in lifter or ' - HW' in lifter or ' - SHW' in lifter:
            curr_quantity = lifter.replace(' - ','')
            dic_lifters[curr_quantity]=[]
        else:
            #Out 代表至少一項沒成績，不用加入
            if 'Out' not in lifter:
                dic_lifters[curr_quantity].append(lifter)

    #過濾掉不需要的資訊並轉成DataFrame
    for key in dic_lifters.keys():
        # print(key)
        ary = None
        for lifer in dic_lifters[key]:

            is_country = 0 
            i=0
            for s in lifer.split(' ')[1:]:
                try:
                    # print(s)
                    if ',' in s:
                        dic_cols[lst_cols[i]].append(float(s.replace(',','.')))
                        i+=1
                    # elif '-' in s: #代表成功後，加重後試舉失敗
                    #     dic_cols[lst_cols[i]].append(0)
                    #     i+=1
                    else:
                        dic_cols[lst_cols[i]].append(float(s))
                        i+=1
                        if is_country == 0:
                            is_country = 1
                except:
                    if is_country == 1:
                        dic_cols[lst_cols[i]].append(s)
                        i+=1
                        is_country=2
                    None
            dic_cols[lst_cols[-1]].append(key)
            
    # for key,val in dic_cols.items():
    #     print(key,len(val),val)

    df = pd.DataFrame(data=dic_cols)
    df['old'] = 2009 - df['old']
    df['old'] = df['old'].astype(int)
    df['year'] = 2009
    if is_men:
        df['gender'] = 'men'
    else:
        df['gender'] = 'women'
    df = df.rename(columns={ col:col.lower() for col in df.columns})
    df = df[['year','wc','gender','old', 'team', 'bwt', 'ipf_gl_cf.', 'sq', 'bp', 'dl', 'total',
        'ipf_gl', 'pts']]
    return df
```

### 將2013的比賽資料轉成 csv檔

```python
def worldGame2013(path:str,is_men:bool)->pd.DataFrame:
    df = pd.read_excel(path)
    df = df.loc[(df['Pts.']!='—')]
    df = df[['BY', 'Nation', 'Weight', 'WF', 'Squat','Bench.', 'Deadlift', 'TOTAL', 'W.Pts.','Pts.']]
    df = df.rename(columns={'BY':'old'})
    lst = df['old'].to_list()
    df['wc'] = 'Light'
    for wc in['Middle','Heavy','SuperHeavy']:
        idx = lst.index(wc)
        df.loc[df.index>=idx,'wc'] = wc
    # lst = [lst.index('Light'),lst.index('Middle'),lst.index('Heavy'),lst.index('SuperHeavy')]
    df = df.loc[(df['old'].isin(['Light','Middle','Heavy','SuperHeavy']))==False]
    df['old'] = 2013 - df['old']
    df['old'] = df['old'].astype(int)
    df['year'] = 2013
    if is_men:
        df['gender'] = 'men'
    else:
        df['gender'] = 'women'
    df = df.rename(columns={ col:col.lower() for col in df.columns})
    df = df[['year','wc','gender','old', 'nation', 'weight', 'wf', 'squat', 'bench.', 'deadlift', 'total','w.pts.', 'pts.']]
    return df
```

### 轉換PDF檔成CSV

```python
import pymupdf
```

```python
def word_game_2017_22_PDF2DataFrame(path,end_word,is_men):
    doc = pymupdf.open(path) # open a document
    page = doc[0]

    #取得線座標
    lst_strikethrough_rect = []
    for item in page.get_drawings():
        rect = item["rect"]
        #判斷刪除線長寬
        a = abs(rect.x0-rect.x1)
        b = abs(458.8800048828125 - 474.7925109863281)
        if (abs(a-b)<=30) & (abs(rect.y1 - rect.y0) <= 1):
                lst_strikethrough_rect.append([rect.x0,rect.y0,rect.x1,rect.y1])
                # print(rect)

    #借用DataFrame特性讓數值能以y0,x0優先順序進行排序，將內容能透過座標按左至右、上到下順序展開
    dic = {key:[] for key in ['x0','y0','x1','y1','val']}
    
    words = page.get_text("words")  # 取得所有單字及座標
    for word in words:
        w = list(word)
        w[4] = w[4].replace(',','.')
        for rect in lst_strikethrough_rect:
            if (abs((w[0]-rect[0]))<=2) and ((w[1]<rect[1]) and (rect[1]<w[3])) and ((w[1]<rect[3]) and (rect[3]<w[3])):
                w[4]='0'
                break
        dic['x0'].append(w[0])
        dic['y0'].append(w[1])
        dic['x1'].append(w[2])
        dic['y1'].append(w[3])
        dic['val'].append(w[4])
    df = pd.DataFrame(dic).sort_values(['y0','x0'],ascending=True)
            
    doc.close()
    df = df.reset_index(drop=True)

    #將展開的值轉成DataFrame
    if '2017' in path:
        lst_cols = 'no. old Team Bwt IPF_GL_Cf. lot SQ1 SQ2 SQ3 SQ_RK BP1 BP2 BP3 BP_RK DL1 DL2 DL3 DL_RK Total IPF_GL WC'.split(" ")
    else:
        lst_cols = 'no. old Team Bwt IPF_GL_Cf. lot SQ1 SQ2 SQ3 SQ_RK BP1 BP2 BP3 BP_RK DL1 DL2 DL3 DL_RK Total IPF_GL Pts WC'.split(" ")
        
    dic = {col:[] for col in lst_cols}

    lst_group = ['Light','Middle','Heavy','SuperHeavy',end_word]

    #分組並濾掉不需要的特殊字元
    dym_col_count = len(lst_cols)-1
    pattern1 = r"[A-Z][a-z]+|[A-Z][a-z]*-[A-Z][a-z]*|[A-Z][a-z]+[A-Z][a-z]+|[A-Z]\.[A-Z]\.|[a-z]+"
    pattern2 = r"\-[cj]|\-[w]|\-[wj]|\-[w][1]|\-[wg]"
    for i in range(len(lst_group)-1):
        d = df.loc[df.loc[df['val']==lst_group[i]].index[0]: df.loc[df['val']==lst_group[i+1]].index[0]-1]
        d = d.loc[d['val'].str.match(pattern1)==False]
        d = d.loc[d['val'].str.match(pattern2)==False]
        d = d.reset_index(drop=True)
        j=0
        
        while(j+dym_col_count<=len(d)):
            for k in range(0,dym_col_count):
                dic[lst_cols[k]].append(d.loc[j]['val'].split("-")[0])
            # print(d[j:j+20]['val'].to_list())
                j+=1   
            dic[lst_cols[-1]].append(lst_group[i])

    # 保留最大的 SQ BP DL 跟去除沒成績的參賽者
    df =  pd.DataFrame(dic)
    for item in ['SQ','BP','DL']:
        lst = [item+str(i)for i in range(1,4)]
        df[item] = df[lst].max(axis=1)
        df = df.drop(columns=lst)
        df = df.loc[((df[item]!='X') & (df[item]!='0') & (df[item]!=0))] 
    df = df.loc[df['Total']!='DD']
    if is_men:
        df['gender'] = 'men'
    else:
        df['gender'] = 'women'
    return df
```

### 重新計算IPF係數
<h> 統一採用2020年5月生效的公式(Equipped Powerlifting)</h>

<h> 參考資料： https://www.powerlifting.sport/rules/codes/info/ipf-formula </h>

<h> 計算方式：</h>

$ \text{IPF GL Coefficient} = \frac{100}{A-B \cdot e^{-C \cdot Bwt} } $



```python
def ipfGLCoefficient_batch(bwt,is_men):
    return [ipfGLCoefficient(bwt[i],is_men[i]) for i in range(0,len(bwt))]
```

```python
dfWorldGame['ipf_gl_c'] = ipfGLCoefficient_batch(dfWorldGame['bwt'].to_list(),(dfWorldGame['gender']==True).to_list())
dfWorldGame
```
